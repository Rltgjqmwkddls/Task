```python
class Person:
    population = 0  # 전체 인구수 (클래스 변수)

    def __init__(self, name):
        self.name = name
        Person.population += 1  # 인스턴스 생성 시 증가
        print(f"{self.name} is born.")

    def __del__(self):
        Person.population -= 1  # 인스턴스 삭제 시 감소
        print(f"{self.name} is dead.")

    @classmethod
    def get_population(cls):
        return cls.population


# 실행 예제
man = Person("james")
woman = Person("emily")
print("전체 인구수: {}명".format(Person.get_population()))

del man
print("전체 인구수: {}명".format(Person.get_population()))
```

```python
class Shop:
    total = 0  # 전체 매출액
    menu_list = [  # 메뉴와 가격
        {"떡볶이": 1000},
        {"순대": 3000},
        {"튀김": 500},
        {"김밥": 2000},
    ]

    @classmethod
    def sales(cls, menu_name, quantity):
        for menu in cls.menu_list:
            if menu_name in menu:  # 메뉴 이름 찾기
                cls.total += menu[menu_name] * quantity
                return

    @classmethod
    def get_total_sales(cls):
        return cls.total


# 실행 예제
Shop.sales("떡볶이", 1)
Shop.sales("순대", 2)
Shop.sales("튀김", 5)
print("총 매출: {}원".format(Shop.get_total_sales()))
```

```python
# 상위 클래스 Car
class Car:
    max_oil = 50  # 최대 주유량

    def __init__(self, oil):
        self.oil = oil

    def add_oil(self, oil):
        if oil < 0:  # 음수는 무시
            return
        self.oil += oil
        if self.oil > Car.max_oil:
            self.oil = Car.max_oil

    def car_info(self):
        print("현재 주유상태: {}".format(self.oil))


# 하위 클래스 Hybrid
class Hybrid(Car):
    max_battery = 30  # 최대 배터리 충전량

    def __init__(self, oil, battery):
        super().__init__(oil)  # 부모 클래스 초기화
        self.battery = battery

    def charge(self, amount):
        if amount < 0:
            return
        self.battery += amount
        if self.battery > Hybrid.max_battery:
            self.battery = Hybrid.max_battery

    def hybrid_info(self):
        print("현재 주유상태: {}, 현재 충전상태: {}".format(self.oil, self.battery))


# 실행 예제
car = Hybrid(0, 0)
car.add_oil(100)
car.charge(50)
car.hybrid_info()
```
