```python
# 사용자로부터 임의의 양의 정수를 하나 입력받은 뒤 1부터 입력받은 정수까지 모든 정수의 합계를 출력
n = int(input('임의의 양수를 입력하세요 '))

t = 0

for n in range(1,n+1):
  t += n
print(f'정수의 합: {t}')
```

```python
# 사용자로부터 임의의 양의 정수를 하나 입력받은 뒤 그 숫자만큼 '과일 이름'을 입력받아 'basket' 리스트에 저장
basket = []

num = int(input('몇 개의 과일을 보관할까요? '))

count = 0

while count < num:
  fruit = input(f'{count+1}번째 과일을 입력하세요 ')
  basket.append(fruit)
  count += 1

print('입력받은 과일들은', basket)
```

```python
# 100점을 받은 학생의 점수를 제외한 나머지 학생들의 점수를 5점씩 증가시킨 'score' 리스트를 생성
exam = [99, 78, 100, 91, 81, 85, 54, 100, 71, 50]

# 새로운 리스트 생성
score = []  # 결과를 저장할 리스트

for x in exam:
    if x < 100:  # 100점이 아닌 경우
        score.append(min(x + 5, 100))  # 5점을 더하고 100점 초과 방지
    else:  # 100점인 경우 그대로 유지
        score.append(x)

# 결과 출력
print("원본 점수:", exam)
print("수정된 점수:", score)
```

```python
# 1부터 99사이의 ㅣ모든 정수를 대상으로 369게임의 결과를 출력
for num in range(1, 100):
    num_str = str(num)  # 숫자를 문자열로 변환
    clap_count = num_str.count('3') + num_str.count('6') + num_str.count('9')  # 3, 6, 9의 개수 계산

    if clap_count > 0:
        print(f"{'짝' * clap_count}", end="\t")  # 박수의 개수만큼 "짝" 출력
    else:
        print(num, end="\t")  # 3, 6, 9가 없으면 숫자 출력

    # 10개씩 한 줄로 출력
    if num % 10 == 0:
        print()  # 줄 바꿈
```
