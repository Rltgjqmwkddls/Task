```python
# 현재 만원을 가지고 있다. 얼마를 사용할 것인지 반복해서 입력받아 만원을 모두 사용
# 0 이하의 금액은 사용불가, 현재 가지고 있는 돈보다 더 큰 금액도 사용불가
coin = 10000
print(f'현재 {coin}원이 있습니다.')

while True:
  n = int(input('사용할 금액 입력: '))

  if n > coin:
    print(f'{coin-n}원이 부족합니다')
  elif n <= 0:
    print('0 이하의 금액은 사용할 수 없습니다.')
  else:
    coin -= n
    print(f'현재 {coin}원이 있습니다.')

  if coin == 0:
    break
```

```python
# 1~5 사이의 정수를 입력받아 평점 만큼 ⭐을 표시
star = '⭐'

while True:
  score = int(input('이번 영화의 평점을 입력하세요: '))

  if 1 <= score <= 5:
    print(score*star)
    break
  else:
    print('평점은 1~5사이만 입력하세요')
```

```python
# 저장된 비밀번호를 맞히는 프로그램
pw = 'dkdlel12'

count = 0

while count <= 5:
  password = input('비밀번호를 입력하세요: ')
  if password != pw:
    print(f'비밀번호를 {count+1}번 불일치합니다.')
    count += 1
    continue
  else:
    print('비밀번호가 일치합니다.')
    break
```

```python
# 짝수인 단(2, 4, 6, 8)은 출력하지 말고 break line만 추가
# 각 단까지만 출력 예, 3단은 3x3까지만출력
dan = 2

while dan <= 9:
    if dan % 2 == 0:  # 짝수 단 건너뛰기
        dan += 1
        continue

    n = 1
    while n <= dan:
        print(f'{dan} * {n} = {dan * n}')
        n += 1

        if n > dan:  # 현재 단의 출력이 끝나면 내부 루프 종료
          print()
          break

    dan += 1  # 다음 단으로 이동
```
