```python
# 정수를 입력받아 그 횟수만큼 'Hello'를 출력
n = int(input("정수를 입력하세요: "))

# 초기화
count = 0

# 입력받은 횟수만큼 'Hello' 출력
while count < n:
    print(f"{count+1}번째 'Hello'")
    count += 1  # 출력 횟수 증가
```

```python
# 1부터 100까지 사이의 모든 정수 중에서 7의 배수만 출력
n = 1 # 7부터 시작
while n <= 100:
  if n % 7 == 0: # 7배수 확인
    print(n)
  n += 1 # 7의 배수만 +
```

```python
# 커피 1잔을 300원에 판매하는 커피자판기에 돈을 넣으면 자판기에서 뽑을 수 있는 커피가 몇 잔이며 잔돈은 얼마인지 출력
coin = int(input('자판기에 얼마를 넣을까요? '))

cup = 0 # 커피 잔

while coin >= 300:
  cup += 1
  coin -= 300
  print(f'커피{cup}잔, 잔돈{coin}원')
```

```python
# 사용자로부터 0부터 9사이의 정수를 입력받아 정수가 5개가 될 때까지 입력받기
s = set()

while len(s) < 5:
  num = int(input('0~9 사이 정수를 입력하세요 '))
  s.add(num)
print(f"입력된 정수: {sorted(s)}")
```

```python
# 1부터 100까지의 모든 정수를 한 줄에 10개씩 출력
n = 0

count = 0

while n < 100:
  n += 1
  print(n, end= ' ')
  count += 1

  if count == 10:
    print()
    count = 0
```

```python
# 전체 구구단을 다음과 같은 모습으로 출력
dan = 2

count = 0

while dan <= 9:

  n = 1
  while n <= 9:
    print(f'{dan}*{n}={dan*n}')
    count += 1
    n += 1
  dan += 1

  if count == 9:
    print()
    count = 0
```
