```python
# 700원짜리 음료수를 뽑을 수 있는 자판기 프로그램 구현
# 돈을 넣으면 몇 잔의 음료수를 뽑을 수 있는지 그리고 잔돈은 얼마인지 모든 경우의 수 출력
def vending_machine(money):
  drink = money // 700
  mon = money % 700
  print('음료수 ={}개, 잔돈= {}원'.format(drink, mon))
vending_machine(3000)
```

```python
# 키가 '과목명', 값이 '점수'인 marks 딕셔너리를 전달하면 해당 딕셔너리에 저장된 점수들의 평균을 반환
def get_average(marks):
  sum = 0
  for i in marks.values(): # marks의 값들만 i에 넣기
    sum += i
  average = sum / len(marks) # 평균 구하기
  return average
marks = {'국어': 90, '영어': 80, '수학': 85}
average = get_average(marks)
print('평균은 {}점입니다.'.format(average))
```

```python
# 결혼식 축의금을 내는 사람의 이름과 그 사람이 낸 축의금을 딕셔너리에 저장
# 전체 축의금의 합계를 구할 수 있는 gift()함수를 구현
total = 0
def gift(dic, who, money):
  wedding = dic
  wedding[who] = money
  global total
  total += money

wedding = {}
gift(wedding, '영희', 5)
gift(wedding, '철수', 3)
gift(wedding, '이모', 10)
print('축의금 명단: {}'.format(wedding))
print('전체 축의금: {}'.format(total))
```
